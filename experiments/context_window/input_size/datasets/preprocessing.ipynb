```python
from transformers import AutoTokenizer
import pandas as pd
import re

csv_files = ["OpenCaselist-100.csv",
             "OpenCaselist-500.csv",
             "OpenCaselist-1000.csv",
             "OpenCaselist-2000.csv",
             "OpenCaselist-3000.csv",
             "OpenCaselist-4000.csv",
             "OpenCaselist-5000.csv",
             "OpenCaselist-10000.csv",
             "OpenCaselist-20000.csv",
             ]
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/pythia-1.4b")

def get_tokens(row):
    return len(tokenizer.encode(f"{row['context']} {row['input']}"))
frames = []
for csv_file in csv_files:
    df = pd.read_csv(f"../datasets/{csv_file}")
    new_df = pd.DataFrame(
        {
            "input": df["fulltext"],
            "context": "summarize the following text: ",
            "expected_output": df["summary"],
            "length": re.split(r'[-.]+', csv_file)[1],
        },
    )

    new_df["total_tokens"] =  new_df.apply(lambda x: get_tokens(x), axis=1)
    frames.append(new_df)

total_df = pd.concat(frames)
total_df
```

```python
df1 = total_df.groupby('length')['total_tokens'].mean()
df2 = total_df.groupby('length')['total_tokens'].std()
calc_df = pd.merge(df1, df2, on='length', how='left')
calc_df = calc_df.rename(columns={'total_tokens_x':'mean', 'total_tokens_y':'std'})
calc_df
```
