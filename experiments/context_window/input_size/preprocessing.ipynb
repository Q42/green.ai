```python
import os
import pandas as pd


csv_files = ["OpenCaselist-100.csv",
             "OpenCaselist-500.csv",
             "OpenCaselist-1000.csv",
             "OpenCaselist-2000.csv",
             "OpenCaselist-3000.csv",
             "OpenCaselist-4000.csv",
             "OpenCaselist-5000.csv",
             "OpenCaselist-10000.csv",
             "OpenCaselist-20000.csv", ]

for csv_file in csv_files:
    df = pd.read_csv(f"../datasets/{csv_file}")
    print(df.head())
    new_df = pd.DataFrame({"input": df["fulltext"], "context": "summarize the following text: ", "expected_output": df["summary"]})
    new_csv_file = f"processed_{csv_file}"
    new_df.to_csv(new_csv_file, index=False)
```
